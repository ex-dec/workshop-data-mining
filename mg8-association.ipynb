```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = pd.read_csv("pembelian.csv")
transaksi = dataset.groupby(['No_Kwitansi','Nama_Barang'])['Jumlah'].sum()
transaksi = transaksi.unstack().reset_index().fillna(0).set_index('No_Kwitansi')
```

```python
transaksi[transaksi>0] = 1
print('Tabel Transaksi \n', transaksi)
```

```python
frequent_itemsets=apriori(transaksi, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
print("\nAssociation Rules: \n", rules[['antecedents', 'consequents', 'confidence']])
```

```python
import pandas as pd

dataset = pd.read_csv("transaction.csv")
display(dataset)
```

```python
data = dataset[dataset["Country"] == "Portugal"]
display(data)
```

```python
transaksi = data.groupby(['InvoiceNo','StockCode'])['Qty'].sum()
transaksi = transaksi.unstack().reset_index().fillna(0).set_index('InvoiceNo')
transaksi[transaksi>0] = 1
display(transaksi)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


frequent_itemsets=apriori(transaksi, min_support=0.2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

print("\nAssociation Rules: \n", rules[['antecedents', 'consequents', 'confidence']])
```
