```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.DataFrame({'val':[66.86,67.47,68.42,69.09,69.58,70.14,70.71,71.29,71.87]})
display(data)
```

```python
x = data.index
y = data.values
plt.scatter(x, y)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
linreg = LinearRegression()
x = np.array(x).reshape(-1,1)
linreg.fit(x, y)
next_x = 9
next_x = np.array(next_x).reshape(-1,1)
pred_x = linreg.predict(next_x)
print('\nPrediksi x \n', pred_x.item())
```

```python
plt.scatter(next_x, pred_x, c='red')
pred_y = linreg.predict(x)
plt.plot(x, pred_y)
plt.show()
```

```python
MSE = mean_squared_error(y,pred_y)
print('\nMSE = ', MSE)
```
